# 문제13
* 주어진 d1에 있는 공원을 아래의 조건에 맞게 folium을 활용하여 지도에 표시하시오
* 1) 공원구분이 '근린공원'이면 color='red', '수변공원'이면 color='green'으로 지정하시오
* 2) 공원구분이 '근린공원', '수변공원'이 아닌 공원은 지도에 표시하지 마시오
* 3) folium.Marker 함수의 tooltip을 사용하고, tooltip시 공원명 나오게 코딩

```python
import folium

d1=pd.read_table('data/dat6.csv', sep=',', encoding='cp949').drop(columns=['공원보유시설(유희시설)','공원보유시설(교양시설)','공원보유시설(기타시설)','관리기관명','제공기관코드','Unnamed: 19'])
d1 = d1.loc[(d1["경도"] < 132) & (d1["위도"] > 32)].copy()
d1=d1.iloc[::600].reset_index(drop=True)
d1.head()
d1['공원구분'].value_counts()


lat = d1["위도"].mean()
long = d1["경도"].mean()

m = folium.Map([lat, long], zoom_start=7, tiles='Stamen Toner')

#=== 코드작성 ===# 

dp=d1[d1['공원구분'].str.contains('근린공원')]
ds=d1[d1['공원구분'].str.contains('수변공원')]

for i in dp.index:
    t1 = dp.loc[i, "공원명"]
    lat = dp.loc[i, "위도"]
    long = dp.loc[i, "경도"]
    
    _=folium.Marker([lat, long], tooltip=t1, radius=3, icon=folium.Icon(color="red")).add_to(m)

for i in ds.index:
    t1 = ds.loc[i, "공원명"]
    lat = ds.loc[i, "위도"]
    long = ds.loc[i, "경도"]
    
    _=folium.Marker([lat, long], tooltip=t1, radius=3, icon=folium.Icon(color="green")).add_to(m)
    
m
```
